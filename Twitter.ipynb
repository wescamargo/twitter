```
#hashtag / dataframe
search_words_dict = ['#COVID19']
execucao = 0
for search_words in search_words_dict:

  cursor_tweets = tw.Cursor(api.search_tweets,q=search_words,lang='pt-br').items(500) #total de itens para pesquisar. Se colocar muito, demora ou toma block
  
  for tweet in cursor_tweets:
    data = {
      'search' :  search_words,
      'text':tweet.text,
      'created_at': tweet.created_at,
      'text': tweet.text, 
      'truncated': tweet.truncated, 
      'source': tweet.source, 
      'user': tweet.user, 
      'is_quote_status': tweet.is_quote_status, 
      'retweet_count': tweet.retweet_count, 
      'favorite_count': tweet.favorite_count,
      'favorited': tweet.favorited, 
      'retweeted': tweet.retweeted,
      'lang': tweet.lang
    }

    df = pd.DataFrame.from_dict(data, orient='index')
    df = df.transpose()

    execucao += 1
    if execucao == 1:
      dftotal = pd.DataFrame.from_dict(data, orient='index')
      dftotal = dftotal.transpose()
    else:
      dftotal = pd.concat([dftotal, df],ignore_index=True)
  
dftotal.head()
```

###transform

```
#clean
dftotal = dftotal.drop(
['truncated',
'is_quote_status',
'user',
'favorited'
], axis =1
)

dftotal.head()
```

```
#rename
dftotal = dftotal.rename(columns={
                                    "search":"Hashtag",
                                    "text":"Texto",	
                                    "created_at":"Criado",	
                                    "source":"Origem",
                                    "retweet_count":"total_retweetado",
                                    "favorite_count":"total_favoritado",
                                    "lang":"Linguagem",
                                    "retweeted":"retweetado"
                                   })
dftotal.head()
```
